## Rank Tensor


```python
#rank 0 tensor  = scalar


rank_0_tensor = tf.constant(15)


print(tf.rank(rank_0_tensor))
print( "Rank of the tensor is " , tf.rank(rank_0_tensor).numpy())
```

```python
# Rank 1 tensor = 1-d array
rank_1_tensor = tf.constant([15 , 12 ])

print(tf.rank(rank_1_tensor))
print( "Rank of the tensor is " , tf.rank(rank_1_tensor).numpy())


print(" Dimension of array " , (rank_1_tensor.numpy()).ndim )
```

## rank 2 tensor   = double [[

```python
#rank 2 tesor
rank_2_tensor = tf.constant([

                             [12 , 13 ] ,
                             [14 , 15 ]

                            ])

```

```python
print(tf.rank(rank_2_tensor))

print()

print( f"This is {tf.rank(rank_2_tensor).numpy()}-D tensor " )
```

```python
#more exmples  of 2-d tensor
a1 = tf.constant([[2 ]])
print(f"{a1.numpy()} is {tf.rank(a1).numpy()}- D tensor ")

```

## Rank 3 tensor

```python
c = tf.constant([[

     [12 , 13  ] ,
     [14 , 15 ]

     ]])

#note that the shape() means this is bad example of tensor
tf.rank(c)
```

```python
d = tf.constant([

                 [

                 [2 , 3 , 6  , 4 ] ,
                 [6  , 4 , 8 , 4  ]

                 ],


                 [[9 , 10 , 11 ,  4 ] ,
                 [12 , 13 , 14   , 4 ]],


                [[4 , 5 , 6  , 4   ] ,
                 [7 , 8 , 9  , 4 ]]


                 ])
```

```python
print(d)
print()
print()

print(f"The given tensor is  {tf.rank(d).numpy()} -d tensor and \n {d.numpy().ndim} -d array of shape {d.numpy().shape}  " )
print()
```

## Reduce

###  *tf*.reduce refers to a set of functions that perform reduction operations on tensors. A reduction operation reduces the dimensions of a tensor by aggregating its values along one or more axes.


#### tf.reduce functions take two main arguments: the input tensor and the axis or axes

######  aggregating -process of combining multiple values into a single value based on a specified operation

```python
print(tf.reduce_max(mat1).numpy())
print(tf.reduce_min(mat1).numpy())
```

```python
print(tf.reduce_min(mat2).numpy())
print(tf.reduce_max(mat2).numpy())
```

```python
# same as numpy
print(np.max(mat1))
print(np.min(mat1))

print()
print(np.max(mat2))
print(np.min(mat2))

```
